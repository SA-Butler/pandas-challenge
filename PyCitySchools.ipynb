```python
# Dependencies and Setup
import pandas as pd
# import numpy as np
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#take the all data, create a new copy
school_data_complete2 = school_data_complete.copy()

school_data_complete2['Per Student Budget'] = school_data_complete2['budget'] / school_data_complete2['size']
                                                                
def math_pass(value):
    if value < 70:
        return 0
    elif value >69:
        return 1
    
school_data_complete2['math_pass'] = school_data_complete2['math_score'].map(math_pass)

def read_pass(value):
    if value < 70:
        return 0
    elif value >69:
        return 1
    
school_data_complete2['read_pass'] = school_data_complete2['reading_score'].map(read_pass)

def both_pass(row):
    if row['math_pass'] == 1 & row['read_pass'] == 1: 
                
        val = 1    
    else:
        val = 0
    return val

school_data_complete2['all_pass'] = school_data_complete2.apply(both_pass, axis=1)

#convert %
school_data_complete2['% Passing Math'] = school_data_complete2['math_pass'] * 100
school_data_complete2['% Passing Reading'] = school_data_complete2['read_pass'] * 100
school_data_complete2['% Overall Passing'] = school_data_complete2['all_pass'] * 100


#Column management 
#remove unwanted columns
school_data_complete2 = school_data_complete2.drop(columns=['read_pass'])
school_data_complete2 = school_data_complete2.drop(columns=['all_pass'])
school_data_complete2 = school_data_complete2.drop(columns=['math_pass'])
school_data_complete2 = school_data_complete2.drop(columns=['Student ID'])
school_data_complete2 = school_data_complete2.drop(columns=['School ID'])


school2 = school_data_complete2.groupby(['school_name','type']).agg('mean').sort_values(by=['budget'],ascending=False)
school2
 
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
school_data_complete3 = school_data_complete.copy()
school_data_complete3.rename(columns = {'School ID':'School_ID'}, inplace = True)
school_data_complete4 = school_data_complete3.copy()

# create list of grades
gradelist = ['9th','10th','11th','12th']
schoollist = ['Huang High School','Figueroa High School','Shelton High School','Hernandez High School','Griffin High School','Wilson High School','Cabrera High School','Bailey High School','Holden High School','Pena High School','Wright High School','Rodriguez High School','Johnson High School','Ford High School','Thomas High School']

sc_summary_df = pd.DataFrame(columns=['School_Name','Grade','Average Math Score'])

# for loop to go through the search lists
for i in schoollist:
   
    for j in gradelist:

            filter1 = school_data_complete4[(school_data_complete4.school_name == i) & (school_data_complete4.grade == j)]
                        
            av_mathscore = filter1['math_score'].mean()
                      
            sc_summary_df.loc[len(sc_summary_df)] = [i, j, av_mathscore]
sc_summary_df
         
```
